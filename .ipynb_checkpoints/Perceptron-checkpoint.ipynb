# Bibliotecas

```python
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import Perceptron
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
import numpy as np
import pandas as pd
import seaborn as sns
```

# Conhecendo o dataset

```python
dados = load_iris()
print(load_iris.__doc__)
```

# Minerando os dados a serem trabalhados

```python
irisDF = pd.DataFrame(data=dados.data, columns=dados.feature_names)
irisDF.tail()
```

```python
irisDF.describe()
```

```python
irisDF.info()
```

```python
objetivos = dict(zip(np.unique(dados.target), dados.target_names))
targetDF = pd.DataFrame.from_dict(objetivos, orient='index', columns=['classes'])
targetDF
```

```python
irisDF['species value'] = pd.Series(dados.target)

species_name = []

for specie in irisDF['species value']:
    if specie == 0:
        species_name.append('setosa')
    elif specie == 1:
        species_name.append('versicolor')
    else:
        species_name.append('virginica')
        
irisDF['species name'] = pd.Series(species_name)

sns.pairplot( data=irisDF, vars=('sepal length (cm)','sepal width (cm)','petal length (cm)','petal width (cm)'), hue='species name' )
```

# Setosa

## Separando as variáveis

```python
#Variaveis independentes
X = dados.data

#Variaveis dependentes em 2 classes: Setosa e não setosa
Y = (dados.target == 0).astype(np.int8)
print(Y)
```

## Treinando o modelo de Perceptron

```python
#separação do conjunto de treino e teste usando 80 treino/20 teste e usando estado randômico a cada execução
X_treino, X_teste, Y_treino, Y_teste = train_test_split(X, Y, test_size = 0.2, random_state=None)
```

```python
#normalizando os dados
norm = StandardScaler()

norm.fit(X_treino)

X_treino_norm = norm.transform(X_treino)
X_teste_norm = norm.transform(X_teste)

X_treino_norm = X_treino_norm[:, [2,3]]
X_teste_norm = X_teste_norm[:, [2,3]]
```

```python
#formato dos dados de treino e teste
print(X_treino.shape, Y_treino.shape)
print(X_teste.shape, Y_teste.shape)
```

```python
modelo = Perceptron(eta0=0.15, max_iter=100, shuffle=False, verbose=1, random_state=1)
```

```python
modelo.fit(X_treino_norm,Y_treino)
```

```python
predicoes = modelo.predict(X_teste_norm)
```

```python
print(Y_teste) #-> valor esperado
print(predicoes) #-> valor predito
```

## Avaliando o modelo 

```python
matriz_confusao = confusion_matrix(Y_teste, predicoes)
print(matriz_confusao)
```

```python
relat_classificacao = classification_report(Y_teste, predicoes)
print(relat_classificacao)
```

```python
acuracia = accuracy_score(Y_teste, predicoes)
print(acuracia)
```
